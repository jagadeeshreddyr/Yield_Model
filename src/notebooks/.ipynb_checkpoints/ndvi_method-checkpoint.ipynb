```python
def ndviImage(image):
    ndvi = image.normalizedDifference(["B8", "B4"])
    return image.addBands(ndvi.rename("NDVI"))

def saviImage(image):
    band_8 = image.select('B8')
    band_4 = image.select('B4')
    numerator = (band_8.subtract(band_4)).multiply(1.5)
    denominator = band_8.add(band_4).add(0.5)
    savi = numerator.divide(denominator)
    return image.addBands(savi.rename("SAVI"))

def lswiImage(image):
    lswi = image.normalizedDifference(["B8A", "B11"])
    lswi = lswi.reproject(crs=lswi.projection().crs(), scale=10)
    lswi = (lswi.subtract(-0.15)).divide(0.50)
    return image.addBands(lswi.rename("LSWI"))

def ndwiImage(image):
    ndwi = image.normalizedDifference(["B3", "B8"])
    return image.addBands(ndwi.rename("NDWI"))

def smooth_6(farm_sep, deg = 6):
    degree = deg  # nth degree polynomial
    x = numpy.array(range(len(farm_sep)))
    y = numpy.array(farm_sep)
    x_test = x
    X = numpy.array([x**i 
                     for i in range(degree + 1)]).T
    X_test = numpy.array([x_test**i 
                          for i in range(degree + 1)]).T
    regr = LinearRegression()

    regr.fit(X, y)

    return regr.predict(X_test)


def ihateclouds(df, k=2):
    series = df["NDVI"].values
    date = df["date"].values
    cloudless = series.copy()
    dfactor = 1

    for i in range(1, len(series)):
        if series[i] <= 0.30:
            index = i 
            while (series[index] <= 0.30 and 
                       ((date[i] 
                         - date[index]).astype('timedelta64[D]')
                        /numpy.timedelta64(1, 'D')) <= 15 and 
                       index >= 0):
                index = index - dfactor
                prev  = series[index]

            index = i + 1
            
            if index < len(series):
                next_val = series[index]
                while (series[index] <= series[i] and 
                           ((date[index] 
                             - date[i]).astype('timedelta64[D]')
                            /numpy.timedelta64(1,'D')) <= 15 and 
                           index < len(series) - 1):
                    index    = index + dfactor
                    next_val = series[index]
                cloudless[i] = (prev 
                                + next_val)/2
            else:
                cloudless[i] = prev

    def onlypeaks(series, k):
        if k == 0:
            return series

        peaks = series.copy()
        for i in range(1, len(series)):
            if ((series[i] 
                     - series[i - 1]) <= 0.5 and 
                    (series[i] 
                     - series[i - 1]) >= 0.01 and 
                    series[i - 1] > 0.3):
                peaks[i - 1] = numpy.nan

        nans, x = numpy.isnan(peaks), lambda z: z.nonzero()[0]
        peaks[nans] = numpy.interp(x(nans), 
                                   x(~nans), 
                                   peaks[~nans])

        return onlypeaks(peaks, k - 1)

    return onlypeaks(cloudless, k)

def get_df(feature):
    coordinates = feature['geometry']['coordinates']
    _id = feature['id']
    dates = feature['properties'].keys()
    ndvi = feature['properties'].values()
    df = pandas.DataFrame()
    df['date'] = dates
    df['NDVI'] = ndvi
    df['id'] = _id
    df['coords'] = "_".join(numpy
                            .array(coordinates)
                            .astype(str))
    return df

get_df = numpy.vectorize(get_df)

# def get_mean_ndvi(AOI, start_date, end_date) :
#     dataset = (ee.ImageCollection("COPERNICUS/S2_SR")
#            .filterBounds(AOI)
#            .filterDate(start_date, end_date)
#            .map(ndviImage)
#            .select("NDVI"))

#     timeseries_image = dataset.toBands()
#     vectors = timeseries_image.sample(region = AOI, 
#                                       scale = 10, 
#                                       geometries = True)
#     features = vectors.getInfo()['features']
#     all_points_df = get_df(numpy.array(features))
    
    
#     final_df = None
#     final_df = pandas.concat(all_points_df, 
#                              ignore_index=True)
#     final_df.date = (final_df.date.str[:8]
#                      .apply(lambda x : datetime.strptime(str(x), "%Y%m%d")))
#     final_df = final_df.loc[final_df.NDVI >= 0 ]
    
#     point_ids = final_df['id'].unique()
#     final_df_group = final_df.groupby('id')

#     final_df_stat = final_df.groupby('date').mean().reset_index()
#     final_df_stat['NDVI_IHC'] = ihateclouds(final_df_stat)
#     final_df_stat['NDVI_SMOOTH'] = smooth_6(final_df_stat['NDVI_IHC'].values, 3)
    
#     return final_df_stat
```

```python
def get_pixel_indices(aoi):    
    dataset = (ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED').filterDate(
        '2022-11-01', '2023-07-01').filterBounds(aoi).map(ndwiImage).map(
            lswiImage).map(ndviImage).map(saviImage))
    savi = (dataset.select('SAVI').toBands()).sample(region=aoi, scale=10, geometries=True)
    ndvi = (dataset.select('NDVI').toBands()).sample(region=aoi, scale=10, geometries=True)
    ndwi = (dataset.select('NDWI').toBands()).sample(region=aoi, scale=10, geometries=True)
    lswi = (dataset.select('LSWI').toBands()).sample(region=aoi, scale=10, geometries=True)
    
    ee_dict = {
        'SAVI': savi.getInfo(),
        'NDVI': ndvi.getInfo(),
        'NDWI': ndwi.getInfo(),
        'LSWI': lswi.getInfo()
    }
    
    return ee_dict
```

```python
def save_csv(image_val, out_sat, farm_id):    
    for index in ['NDVI', 'SAVI', 'LSWI', 'NDWI']:
        features = image_val[index]['features']
        prop = [
            pd.DataFrame(list(
                zip(list(features[x]['properties'].keys()),
                    list(features[x]['properties'].values()))),
                         columns=['date', x]) for x in range(len(features))
        ]
        final_df = prop[0]
        for df in prop[1:]:
            final_df = final_df.merge(df, on='date')
        final_df['date'] = final_df['date'].str[:8]
        final_df.to_csv(f"{out_sat}{farm_id}_{index}.csv", index=False)
```

```python
def inter_pol( df,days = 130):
    d = []
    y_interp = interp1d(df['days'], df['NDVI_SMOOTH'])
    date = df['date'][0]
    
    for i in range(days):
        date += timedelta(days= 1)
        d.append((date, float(y_interp(i)), i))
        
    return pandas.DataFrame(d, columns = ['date', 'NDVI_SMOOTH','days'])
```
